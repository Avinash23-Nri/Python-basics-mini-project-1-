```python
l1= [10,29,61,7,20,9,11,1,5]

# Create an empty list to store sorted elements
l2= []


for i in range(len(l1)):
    min_val = l1[0]
    for j in range(1, len(l1)):
        if l1[j] < min_val:
            min_val = l1[j]

    l2.append(min_val)

    l1.remove(min_val)

# Print the sorted list
print("Sorted list in ascending order:", l2)
```
